```python
#import library pandas & generate data dummy, lalu tampilkan pada DataFrame
import pandas as pd

data = {'luas' : [6, 8, 10, 14, 18],
        'harga' : [7, 9, 13, 17.5, 18]}

df = pd.DataFrame(data)
df
```

```python
#Visual data dengan scatterplot menggunakan library matplotlib
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='luas', y='harga')
plt.title('Perbandingan Luas Tanah Dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
#Memisahkan variabel dependen dengan variabel independent
#Variabel y=label/target & Variabel X=fitur
import numpy as np

X = np.array(df['luas']).reshape(-1, 1)
y = np.array(df['harga'])
```

```python
#Latih model (fit)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)

print(f'intercept : {model.intercept_}')
print(f'slope : {model.coef_}')
```

```python
#Visualisasi garis persamaan linear dihasilkan model setelah dilatih
X_vis = np.array([0, 25]).reshape(-1,1)
y_vis = model.predict(X_vis)

plt.scatter(X, y)
plt.plot(X_vis, y_vis, color='red')

plt.title('Perbandingan Luas Tanah Dengan Harga Tanah')
plt.xlabel('Luas Tanah (m persegi)')
plt.ylabel('Harga Tanah (dollar)')
# plt.xlim(0, 25)
# plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

```python
#Prediksi harga rumah untuk tanah 12, 15, dan 20 m
luas_tanah = np.array([12,15,20]).reshape(-1,1)
luas_tanah

harga_prediksi = model.predict(luas_tanah)
harga_prediksi
```

```python
#Tampilkan hasil prediksi harga rumah
for luas, hrg in zip(luas_tanah, harga_prediksi):
    print(f'Luas Tanah {luas}: harga prediksi: {hrg}')
```
