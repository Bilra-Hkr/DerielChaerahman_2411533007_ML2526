```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set_style("darkgrid")
```

```python
# Load Import dataset from github
dataset_url = 'https://raw.githubusercontent.com/Bilra-Hkr/DerielChaerahman_2411533007_ML2526/refs/heads/main/Praktikum2/Breast_cancer_data.csv'
data = pd.read_csv(dataset_url)
data.head()
```

```python
data.tail()
```

```python
# Visualisasi data Histogram menggunakan library seaborn
data["diagnosis"].hist()
```

```python
# Prepare features (X) and target (y)
X, y = data.drop(columns='diagnosis'), data['diagnosis']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and train the Gaussian Naive Bayes model
sklearn_gnb = GaussianNB()
sklearn_gnb.fit(X_train, y_train)

# Make predictions on the test set
y_pred = sklearn_gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay

print('Sklearn Akurasi = ', accuracy_score(y_test, y_pred))
print('Sklearn F1-Score = ', f1_score(y_test, y_pred))

# Display confusion matrix
cm_sklearn = confusion_matrix(y_test, y_pred)
disp_sklearn = ConfusionMatrixDisplay(confusion_matrix=cm_sklearn, display_labels=labels)
disp_sklearn.plot();
plt.title('Confusion Matrix (Sklearn Gaussian Naive Bayes)');
```
